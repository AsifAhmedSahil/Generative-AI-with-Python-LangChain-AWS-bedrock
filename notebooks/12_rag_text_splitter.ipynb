### Documents Text Splitters

```python
from langchain_community.document_loaders import TextLoader,PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
loader = TextLoader("../data/speech.txt",encoding="utf-8")
texts = loader.load()
```

```python
loader = PyPDFLoader("../data/medical_report.pdf")
pdfs = loader.load()
```

```python
splitters = RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=200)
```

```python
text_docs = splitters.split_documents(texts)
print(text_docs)
```

```python
pdf_docs = splitters.split_documents(pdfs)
len(pdf_docs)
pdf_docs[0]
```
