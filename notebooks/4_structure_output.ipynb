```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model="gpt-4o-mini")
```

### Discuss the problem first

```python
text = "Hello, my name is asif ahmed sahil ,Email is asif@gmail.com and my age is 25"
```

```python
res = llm.invoke(f"Please give me only name,email and age from the text:{text}")
```

```python
print(res.content)
print(type(res.content))
```

### Solution is : Structured output

```python
from pydantic import BaseModel,Field
from typing import List

class ResponseStructure(BaseModel):
    name:str = Field(description="Complete Name")
    email: str  = Field(description="Email address")
    age: int = Field(description="Age")

structured_llm = llm.with_structured_output(ResponseStructure)

```

```python
res = structured_llm.invoke(f"Please give me only name,email and age from the text:{text}")
```

```python
res.name
```

```python
class Moveis(BaseModel):
    title:str = Field(description="Movie title")
    year:int = Field(description="Movie release year")
class AllMovies(BaseModel):
    movies:List[Moveis]

movie_llm = llm.with_structured_output(AllMovies)
```

```python
res = movie_llm.invoke("Give me three sci-fi movie")
```
