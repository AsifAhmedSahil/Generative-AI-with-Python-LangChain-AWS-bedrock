```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_core.tools import tool
```

```python
@tool
def add_number(a:int, b:int):
    """ 
        it will return the sum of two numbers

        Args:
            a: Number One
            b: Number Two
    """
    return a+b
@tool
def multiply_number(a:int, b:int):
    """ 
        it will return the product of two numbers

        Args:
            a: Number One
            b: Number Two
    """
    return a*b
```

```python
from langchain_openai import ChatOpenAI
from langchain.agents import create_agent
```

```python
llm = ChatOpenAI(model="gpt-4o-mini")
```

```python
agent = create_agent(
    model=llm,
    tools=[add_number,multiply_number],
    system_prompt="You are a math teacher and always use tools for calculation"
)
```

```python
response = agent.invoke({"messages":[{"role":"user","content":"what is 2+3*8?"}]})
print(response["messages"][-1].content)
```
