### Text Document Loader

```python
from langchain_community.document_loaders import TextLoader
loader = TextLoader("../data/speech.txt",encoding="utf-8")
texts = loader.load()

onedoc = texts[0]
print(onedoc.page_content)
```

### PDF Document loader

```python
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader("../data/medical_report.pdf")
docs = loader.load() 
len(docs)
```

```python
onedoc = docs[0]
print((onedoc.page_content))
```

```python
from langchain_community.document_loaders import CSVLoader
loader = CSVLoader("../data/employee_records.csv")
docs = loader.load()
docs[:5]
```

### Web based document load

```python
from langchain_community.document_loaders import WebBaseLoader
loader = WebBaseLoader(web_path="https://www.becodemy.com/#pricing")
docs = loader.load()
print(docs[0].page_content)
```

### Wikipedia loader

```python
from langchain_community.document_loaders import WikipediaLoader
loader = WikipediaLoader(query="Gen ai",load_max_docs=2)
docs = loader.load()
docs[0].page_content
```
