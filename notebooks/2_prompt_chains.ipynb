### Prompt Chaining

```python
from langchain_openai import ChatOpenAI
```

```python
llm = ChatOpenAI(model="gpt-4o-mini")
```

```python
prompts = [
    {"role":"system","content":"you are a js developer"},
    {"role":"user","content":"tell me map in js?"}
 ]

res = llm.invoke(prompts)
print( res.content)
```

#### Prompt Template - Dynamic prompts

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
llm = ChatOpenAI(model="gpt-4o-mini")
out = StrOutputParser()
```

```python
prompts = ChatPromptTemplate.from_messages([
    {"role":"system","content":"you are a {language} developer"},
    {"role":"user","content":"{query}"}
 ])
```

```python
final_prompts = prompts.format_messages(language ="javascript",query = "tell me map in js?")
res = llm.invoke(final_prompts)
print(res.content)
```

### Chains - Runnables - jake run kora jai mane invoke kora jai 

```python
res = llm.invoke(prompts.invoke({"language":"python","query":"what is python?"}))
res.content
```

### chain function - uppercase convert custom function

```python
def transform_case(result:str):
    return result.upper()
```

### aikhane chain kora hoise mane  prompts k invoke kore j output pawa jabe seta k llm e invoke kroa hoi then ans generate hoi 

### same jinish chain dia kora jai short cut - chains o same runable er moto kaj krobe

```python
chains = prompts | llm | out | transform_case
```

```python
res = chains.invoke({"language":"python","query":"what is python?"})
res
```
